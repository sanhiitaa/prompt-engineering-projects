```python
from manim import *

class L2Normalization(Scene):
    def construct(self):
        title = Text("L2 Normalization", font_size=48)
        title.to_edge(UP)

        # Equation of L2 Norm
        l2_norm_eq = MathTex(r"\| \mathbf{w} \|_2 = \sqrt{w_1^2 + w_2^2 + \cdots + w_n^2}", font_size=36)
        l2_norm_eq.next_to(title, DOWN, buff=0.5)

        # Add title and equation to the scene
        self.play(Write(title))
        self.play(Write(l2_norm_eq))
        self.wait(2)

class RidgeRegression(Scene):
    def construct(self):
        title = Text("Ridge Regression", font_size=48)
        title.to_edge(UP)

        # Ordinary Least Squares equation
        ols_eq = MathTex(r"\text{OLS: } \min_{\mathbf{w}} \| \mathbf{Xw} - \mathbf{y} \|^2", font_size=36)
        ols_eq.next_to(title, DOWN, buff=0.5)

        # Ridge Regression equation
        ridge_eq = MathTex(r"\text{Ridge: } \min_{\mathbf{w}} \| \mathbf{Xw} - \mathbf{y} \|^2 + \lambda \| \mathbf{w} \|_2^2", font_size=36)
        ridge_eq.next_to(ols_eq, DOWN, buff=0.5)

        explanation = Text("""
        Ridge Regression adds a penalty term to the OLS loss function
        to prevent overfitting by shrinking the coefficients.
        """, font_size=24)
        explanation.next_to(ridge_eq, DOWN, buff=0.5)

        # Add title, equations, and explanation to the scene
        self.play(Write(title))
        self.play(Write(ols_eq))
        self.play(Write(ridge_eq))
        self.wait(2)
        self.play(Write(explanation))
        self.wait(2)
```

```python
from manim import *

class RidgeRegressionVisualization(Scene):
    def construct(self):
        # Title
        title = Text("Ridge Regression").to_edge(UP)
        des = Text("Ridge regression is a linear regression technique that includes an L2 regularization term to prevent overfitting by penalizing large coefficients.", font_size=20)

        self.play(Write(title))
        self.add(des)
        self.wait(4)
        self.clear()

        # Scene 1: Underfit Model
        self.show_underfit_model()

        # Scene 2: Overfit Model
        self.show_overfit_model()

        # Scene 3: Ridge Regression Model
        self.show_ridge_regression_model()

    def show_underfit_model(self):
        underfit_title = Text("Underfit Model").scale(0.75).to_edge(UP)
        self.play(Write(underfit_title))

        # Example underfit model: a straight line
        axes = Axes(
            x_range=[-1, 10, 1], y_range=[-1, 10, 1],
            axis_config={"include_numbers": True}
        )
        underfit_line = axes.plot(lambda x: 0.5 * x + 1, color=BLUE)

        self.play(Create(axes), Create(underfit_line))
        self.wait(2)
        self.clear()

    def show_overfit_model(self):
        overfit_title = Text("Overfit Model").scale(0.75).to_edge(UP)
        self.play(Write(overfit_title))

        # Example overfit model: a high-degree polynomial
        axes = Axes(
            x_range=[-1, 10, 1], y_range=[-1, 10, 1],
            axis_config={"include_numbers": True}
        )
        overfit_curve = axes.plot(lambda x: 0.1 * (x - 1) * (x - 2) * (x - 3) * (x - 4) + 5, color=RED)

        self.play(Create(axes), Create(overfit_curve))
        self.wait(2)
        self.clear()

    def show_ridge_regression_model(self):
        ridge_title = Text("Ridge Regression Model").scale(0.75).to_edge(UP)
        self.play(Write(ridge_title))

        # Example ridge regression model: a balanced curve
        axes = Axes(
            x_range=[-1, 10, 1], y_range=[-1, 10, 1],
            axis_config={"include_numbers": True}
        )
        ridge_curve = axes.plot(lambda x: 0.05 * (x - 1) * (x - 2) * (x - 3) + 2, color=GREEN)

        self.play(Create(axes), Create(ridge_curve))
        self.wait(2)

        # Conclude with a message
        conclusion = Text("Ridge Regression Generalizes Better").scale(0.75).to_edge(DOWN)
        self.play(Write(conclusion))
        self.wait(2)
        self.clear()

if __name__ == "__main__":
    from manim import config
    config.media_width = "100%"
    scene = RidgeRegressionVisualization()
    scene.render()
```
